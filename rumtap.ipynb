```python
import pandas as pd

data = pd.read_csv('sikumbang.csv')
data.head()
```

TAHAPAN PREPROCESSING

```python
#bersihkan kolom harga
data['harga'] = data['harga'].str.replace(r'[^0-9]', '', regex=True)

# Convert to numeric, setting errors='coerce' to handle non-convertible entries
data['harga'] = pd.to_numeric(data['harga'], errors='coerce')

# Optionally, you can fill NaN values if needed or drop them
# For example, to fill NaN with 0:
data['harga'] = data['harga'].fillna(0).astype(float)
data.head()
```

```python
# Filter out rows where 'harga' is greater than 500,000,000
data = data[data['harga'] <= 500000000]
```

```python
#bersihkan kolom LB dan LL
# Extract numeric values from 'LB' while retaining NaN for non-numeric entries
data['LB'] = data['LB'].str.extract(r'(\d+)')
# Convert extracted values to numeric, keeping NaN values
data['LB'] = pd.to_numeric(data['LB'], errors='coerce')

# Extract numeric values from 'LL' while retaining NaN for non-numeric entries
data['LL'] = data['LL'].str.extract(r'(\d+)')
# Convert extracted values to numeric, keeping NaN values
data['LL'] = pd.to_numeric(data['LL'], errors='coerce')

data.head()
```

```python
#bersihkan data kamar_tidur dan kamar_mandi
data['kamar_tidur'] = data['kamar_tidur'].str.extract(r'(\d+)')
data['kamar_mandi'] = data['kamar_mandi'].str.extract(r'(\d+)')

# Convert extracted values to numeric, keeping NaN values
data['kamar_tidur'] = pd.to_numeric(data['kamar_tidur'], errors='coerce')
data['kamar_mandi'] = pd.to_numeric(data['kamar_mandi'], errors='coerce')

# Convert to integers, if desired, using fillna() to handle NaNs
data['kamar_tidur'] = data['kamar_tidur'].fillna(0).astype(int)
data['kamar_mandi'] = data['kamar_mandi'].fillna(0).astype(int)
data.head()
```

```python
#pastikan semua data dalam lowercase
data = data.applymap(lambda x: x.lower() if isinstance(x, str) else x)
data.head()
```

```python
#hapus semua extra spaces
data = data.applymap(lambda x: ' '.join(x.lower().strip().split()) if isinstance(x, str) else x)
data.head()
```

```python
#bersihkan kolom telp dan email
data['telp'] = data['telp'].str.replace(r'telp :\s*', '', regex=True).where(data['telp'].notna())
data['email'] = data['email'].str.replace(r'email :\s*', '', regex=True).where(data['email'].notna())
data.head()
```

```python
#bersihkan kolom yang lain secara bersamaan pov: capek bet
# Function to clean the columns
def clean_column(column, prefix):
    return column.str.replace(fr'{prefix}\s*', '', regex=True)

# Clean each specified column
data['atap'] = clean_column(data['atap'], r'a\. atap').where(data['dinding'].notna())
data['dinding'] = clean_column(data['dinding'], r'b\. dinding').where(data['dinding'].notna())
data['lantai'] = clean_column(data['lantai'], r'c\. lantai & pondasi').where(data['lantai'].notna())
data['unit_subsidi'] = clean_column(data['unit_subsidi'], r'subsidi\s*:').where(data['unit_subsidi'].notna())
data['unit_komersil'] = clean_column(data['unit_komersil'], r'komersil\s*:').where(data['unit_komersil'].notna())
data.head()
```

```python
#pisahkan latitude dan longitude
# Split the 'longlat' column into 'latitude' and 'longitude' columns with 'n=1' to limit to one split
data[['latitude', 'longitude']] = data['longlat'].str.split(',', n=1, expand=True)

# Convert columns to numeric, with errors='coerce' to handle any non-numeric values
data['latitude'] = pd.to_numeric(data['latitude'], errors='coerce')
data['longitude'] = pd.to_numeric(data['longitude'], errors='coerce')

# Drop the original 'longlat' column if it's no longer needed
data = data.drop(columns=['longlat'])

# Display the updated DataFrame
data.head()
```

```python
#ambil alamat kabupaten atau kota saja
import pandas as pd
import numpy as np
import re
# Function to extract city/regency name
def extract_kabupaten_kota(location):
    if pd.isna(location):
        return np.nan
    match = re.search(r', ([^,]+),', location)  # Match the city/regency name
    return match.group(1).strip() if match else np.nan

# Apply the function to the column
data['kabupaten_kota'] = data['kabupaten_kota'].apply(extract_kabupaten_kota)

data.head()
```

TAHAPAN PREPROCESSING

```python
#variable yang menyimpan kolom string/object
kol_kategori = [kol for kol in x_train.columns 
                if x_train[kol].nunique() < 30
                and x_train[kol].dtype == "object"]
```

```python
#Variable yang tipenya numerik (int atau float)
kol_numerik = [kol for kol in x_train.columns
               if x_train[kol].dtype in ['int64', 'float64']]
```

```python
#import library yang dibutuhkan
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

#variable untuk menyimpan langkah 'imputer' (merubah entri null menjadi mean) pada kolom numerik
# Update numerical preprocessing step to include scaling
kol_ubah_num = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

#variable untuk menyimpan langkah 'imputer dan Ordinal Encoding' (merubah entri null menjadi modus dan merubah kolom kategori menjadi integer)
#kol_ubah_kat = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),
#                               ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
#                              ])
# Update pipeline for categorical features to use OneHotEncoder
kol_ubah_kat = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

#pastikan setiap unique entry di x_train juga ada di x_test
#entry_aman = [kol for kol in kol_kategori if set(x_test[kol]).issubset(set(x_train[kol]))]
# Update entry_aman to include atap and dinding explicitly

#terapkan langkah daiatas ke kolom dataframe
preprocessor = ColumnTransformer(transformers=[('num', kol_ubah_num, kol_numerik),
                                               ('kat', kol_ubah_kat, kol_kategori)
                                              ])
```

TAHAP PEMODELAN

```python
#definisikan model
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=200, random_state=0)

#satukan tahap preprocessing dan model
pipeline = Pipeline(steps=[('clear_data', preprocessor),('model', model)])

#fit kan model dengan x_train dan y_train
pipeline.fit(x_train, y_train)
```

TAHAP PREDIKSI DAN EVALUASI MODEL

```python
#prediksi model
predik = pipeline.predict(x_test)

print(predik)
```
